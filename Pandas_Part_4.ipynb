```python
df7=pd.DataFrame({'date':['2022-06-22','2022-06-20']})
```

```python
df7
```

```python
df7.dtypes
```

```python
df7['year']=df7['update_date'].dt.year # Extract Year out of date
```

```python
df7
```

```python
df7['month']=df7['update_date'].dt.month #Extract month out of date
```

```python
df7
```

```python
df7['day']=df7['update_date'].dt.day #Extract day out of date
```
