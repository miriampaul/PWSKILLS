```python
data1=pd.DataFrame({'key1':[1,2,56],
                    'key2':[24,54,343],
                    'key3':[3,4343,554]})
```

```python
data2=pd.DataFrame({'key3':[3,433,554],
                    'key4':[24,5544,343],
                    'key5':[334,4343,554]})
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1, data2) # Inner join
```

```python
#Inner join - Matching rows in both tables or common rows 
# HEre key3 is a common column name , respective values will be recheck 
```

```python
pd.merge(data1,data2, how='left')
```

```python
pd.merge(data1, data2, how='right')
```

```python
pd.merge(data1, data2, how='outer')
```

```python
pd.merge(data1, data2, how='cross')
```

```python
pd.merge(data1,data2,how='left',on='key3')
```

```python
#If both tables column names are different then use Right_on or Left_on
```

```python
pd.merge(data1, data2, how='left', right_on='key4', left_on='key2')
```
